### 1. Data

```python
def normalize_data(df):
    '''
    Use the preprocessing object MinMaxScaler from sklearn library to normalize the data.
    The out dataframe has the same index and column
    '''
    x = df.values 
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df_n = pd.DataFrame(x_scaled, index=df.index, columns=df.columns)
    return df_n
```

```python
# read csv
stock_last_price = pd.read_csv('stock.csv', index_col=0) # data ranging from 2016-1-4 to 2020-6-4
stock_pct = pd.read_csv('stock_pct.csv', index_col=0)
index = pd.read_csv('sp500_index.csv', index_col=0) 

index_last_price = index['Adj Close']
index_pct = index['Adj Close'].pct_change()

# # stock data
# s_lp_calibrate = stock_last_price.iloc[:780, :]
# s_lp_validate = stock_last_price.iloc[780:, :]

# # s&p 500 data
# i_lp_calibrate = pd.Series(index_last_price)[:780]
# i_lp_validate = pd.Series(index_last_price)[780:]
stock_last_price.index= pd.to_datetime(stock_last_price.index)
```

```python
list_tickers = stock_last_price.columns # array type
dates = stock_last_price.index.values # array type
#stock_last_price.index= pd.to_datetime(stock_last_price.index) 
stock_last_price_n = normalize_data(stock_last_price) # normalized adj close
```

### 2. Deep AE

```python
encoding_dim = 15 # 15 neurons
num_stock = len(stock_pct.columns) # Use 481 stocks as features

# connect all layers
# add more hidden layer, use Adam as optimizer, use normalized adj close as data
input_img = Input(shape=(num_stock, ))
encoded = Dropout(0.2)(input_img)
encoded = Dense(200,          activation='relu', kernel_constraint=maxnorm(3))(encoded)
encoded = Dense(encoding_dim, activation='relu', kernel_constraint=maxnorm(3))(encoded)
decoded = Dense(200,          activation= 'relu', kernel_constraint=maxnorm(3))(encoded) 
decoded = Dense(num_stock,    activation= 'linear')(decoded)

# construct and compile AE model
autoencoder = Model(input_img, decoded)
from keras.optimizers import SGD
sgd = SGD(lr=0.1, momentum=0.9, decay=0.0, nesterov=False)
autoencoder.compile(optimizer=sgd, loss='mean_squared_error')

# train autoencoder
#data = s_lp_calibrate.diff()[1:]
train_data = stock_last_price_n.iloc[:780,]
valid_data = stock_last_price_n.iloc[780:,]
train_autoencoder = autoencoder.fit(train_data, train_data, shuffle=False, epochs=100, batch_size = 10, validation_data=(valid_data, valid_data))
autoencoder.save('model/retrack_autoencoder.h5')

# test/reconstruct market information matrix
reconstruct = autoencoder.predict(train_data)
```

```python
loss = train_autoencoder.history['loss']
val_loss = train_autoencoder.history['val_loss']
epochs = range(100)
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

```python
communal_information = []

for i in range(0,481):
    diff = np.linalg.norm((train_data.iloc[:,i] - reconstruct[:,i])) # 2 norm difference
    communal_information.append(float(diff))
 
print("stock #, 2-norm, stock name")
ranking = np.array(communal_information).argsort()
for stock_index in ranking:
    print(stock_index, communal_information[stock_index], stock_last_price.iloc[:,stock_index].name) # print stock name from lowest different to highest
```

```python
# stock data
s_lp_calibrate = stock_last_price.iloc[:780, :]
s_lp_validate = stock_last_price.iloc[780:, :]
s_p_calibrate = stock_pct.iloc[0:780, :]
s_p_validate= stock_pct.iloc[780:, :]

# s&p 500 data
i_lp_calibrate = pd.Series(index_last_price)[:780]
i_lp_validate = pd.Series(index_last_price)[780:]
i_p_calibrate = pd.Series(index_pct)[:780]
i_p_validate= pd.Series(index_pct)[780:]
```

```python
# To beat the S&P500 index, we will amend the target data during the calibrate phase by replacing all returns smaller than -5% by exactly 5%
# Aims to create a index tracker with anti-correlation in periods of large drawdowns.

# from -5% to 5%
y_amended = i_p_calibrate
y_amended[y_amended < -0.04] = 0.04

# re-calculate the last price
y_amended[0] = 0
relative_percentage = y_amended + 1
lp_amended = i_lp_calibrate[0] * (relative_percentage.cumprod()) 

# plot comparison
#pd.Series(ibb['calibrate']['lp'].as_matrix(), index=pd.date_range(start='01/06/2012', periods=104, freq='W')).plot(label='ibb original', legend=True) 
#pd.Series(lp_amended.as_matrix(), index=pd.date_range(start='01/06/2012', periods=104, freq='W')).plot(label='ibb amended', legend=True)
plt.figure(figsize=(10,5))
plt.plot(s_lp_calibrate.index, pd.Series(index_last_price)[:780], label='S&P original')
plt.plot(s_lp_calibrate.index, lp_amended, label='S&P amended')
plt.legend()
```

```python
relative_percentage_index = pd.Series(index_pct)[:780] + 1
lp_index = 1 * (relative_percentage_index.cumprod())


y_amended = i_p_calibrate
y_amended[y_amended < -0.04] = 0.04
# re-calculate the last price
y_amended[0] = 0
relative_percentage_index_amended = y_amended + 1
lp_amended = 1 * (relative_percentage_index_amended.cumprod())

plt.figure(figsize=(10,5))
plt.plot(s_lp_calibrate.index, lp_index, label='S&P original')
plt.plot(s_lp_calibrate.index, lp_amended, label='S&P amended')
plt.legend()
```

```python
ibb_predict = defaultdict(defaultdict)
total_2_norm_diff = defaultdict(defaultdict)
dl_scaler = defaultdict(StandardScaler)

for non_communal in [10, 15, 20, 25, 30, 35, 40]:  
    # some numerical values
    encoding_dim = 5
    s = 10 + non_communal
    stock_index = np.concatenate((ranking[0:10], ranking[-non_communal:])) # portfolio index
    
    
    # connect all layers
    input_img = Input(shape=(s,))
    encoded = Dense(encoding_dim, activation='relu', kernel_regularizer=regularizers.l2(0.002))(input_img)
    decoded = Dense(1, activation= 'linear', kernel_regularizer=regularizers.l2(0.002))(encoded)
    
    
    # construct and compile deep learning routine
    deep_learner = Model(input_img, decoded)
    deep_learner.compile(optimizer='sgd', loss='mean_squared_error')
    
    x = s_p_calibrate.iloc[1:,stock_index]
    y = y_amended[1:] # amended percentage
    
    dl_scaler[s] = StandardScaler()       # Multi-layer Perceptron is sensitive to feature scaling, so it is highly recommended to scale your data
    dl_scaler[s].fit(x)
    x = dl_scaler[s].transform(x)  
    
    deep_learner.fit(x, y, shuffle=False, epochs=400, batch_size = 10)    # fit the model
    deep_learner.save('model/beat_s' + str(s) + '.h5') # for validation phase use
    
    
    # is it good?
    relative_percentage = copy.deepcopy(deep_learner.predict(x))
    #relative_percentage[0] = 0
    relative_percentage = relative_percentage + 1
    
    index_predict['calibrate'][s] = i_lp_calibrate[0] * (relative_percentage.cumprod())          
    total_2_norm_diff['calibrate'][s] = np.linalg.norm((index_predict['calibrate'][s] - lp_amended[1:])) # compare with amended last price
```
